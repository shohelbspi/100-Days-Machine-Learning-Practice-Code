# opening a csv file from an url

```python
import requests
from io import StringIO
```

# convertors parameter

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
```

```python

pd.read_csv('datasets/IPL Matches 2008-2020.csv').head(2)
```

```python
pd.read_csv('datasets/IPL Matches 2008-2020.csv',converters={'team1':rename}).head(2)
```
