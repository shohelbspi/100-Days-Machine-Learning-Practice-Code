# Defination Of Normalization?


***Normalization is a technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of numeric columns in the datasetto use a common scale, without distorting differences in the range of values or losing information***

# Type Of Normalization

### MinMax Scaling
### Mean Normalization
### Max Absoulte
### Robust Scaling

# MinMax Scaling 

## Formula
***xi' = xi-xmin/xmax-xmin***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/wine_data.csv',usecols=[0,1,2],header=None)
```

```python
df.head(2)
```

```python
len(df)
```

```python
df.columns=['class_label','Alcohol','Malic Acid']
```

```python
df.head()
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic Acid'])
plt.show()
```

```python
df['class_label'].value_counts()
```

```python
color_dict={1:'red',3:'green',2:'blue'}
sns.scatterplot(x=df['Alcohol'],y=df['Malic Acid'],hue=df['class_label'],palette=color_dict)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('class_label',axis=1)
y = df['class_label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
scaler = MinMaxScaler()

x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scale, columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scale, columns=x_test.columns)
```

```python
x_train_scaled
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))
ax1.set_title('Before Scaling')
ax1.scatter(x_train['Alcohol'],x_train['Malic Acid'],c=y_train)
ax2.set_title('After Scaling')
ax2.scatter(x_train_scaled['Alcohol'],x_train_scaled['Malic Acid'],c=y_train)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'],ax=ax1)
sns.kdeplot(x_train['Malic Acid'],ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(x_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(x_train_scaled['Malic Acid'],ax=ax2)
plt.show()
```
