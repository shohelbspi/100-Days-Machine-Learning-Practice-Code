```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('purchased',axis=1),df['purchased'],test_size=0.23)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()

x_train_encode = oe.fit_transform(x_train)
x_test_encode = oe.transform(x_test)
```

```python
x_train_encode
```

```python
oe2 = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])

x_train_encode2 = oe2.fit_transform(x_train)
x_test_encode2 = oe2.transform(x_test)
```

```python
x_train_encode2
```
